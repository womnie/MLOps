```python
## 1. Import Libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## 2. Load the Dataset
iris = sns.load_dataset('iris')
```

```python
## 3. Data Exploration
# 
iris.head()
iris.describe()
# visualisasikan distribusi panjang dan lebar petal menggunakan histogram
plt.figure(figsize=(10, 6))
sns.histplot(iris['petal_length'], kde=True)
plt.title('Distribution of Petal Length')
plt.show()

# visualisasikan hubungan antara panjang dan lebar petal dengan scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='petal_length', y='petal_width', data=iris, hue='species')
plt.title('Petal Length vs Petal Width by Species')
plt.show()

# lihat korelasi antara fitur menggunakan heatmap:
numeric_cols = iris.select_dtypes(include=[np.number])
# visualisasikan korelasi antara fitur numerik
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Iris Dataset')
plt.show()
```

```python
## 4. Conclusion

#Distribusi Panjang dan Lebar Petal:
#Histogram menunjukkan bahwa distribusi panjang petal mengikuti pola bimodal dengan dua puncak utama, 
#yang mengindikasikan adanya dua kelompok berbeda dalam dataset.
#Distribusi lebar petal juga menunjukkan variasi yang signifikan antar spesies.

#Hubungan Antar Variabel:
#Scatter plot yang memvisualisasikan hubungan antara panjang dan lebar petal menunjukkan bahwa spesies Setosa 
#memiliki ukuran petal yang lebih kecil dan terpisah dengan jelas 
#dari dua spesies lainnya (Versicolor dan Virginica).
#Spesies Versicolor dan Virginica cenderung memiliki ukuran petal yang lebih besar dengan sedikit tumpang tindih, 
#namun masih dapat dibedakan berdasarkan ukurannya.

#Korelasi Antar Fitur:
#Heatmap dari matriks korelasi mengungkapkan bahwa ada korelasi positif yang kuat antara panjang dan lebar petal (petal_length dan petal_width), 
#yang menunjukkan bahwa semakin panjang petal, semakin lebar pula ukurannya.
#Sebaliknya, korelasi antara panjang sepal (sepal_length) dan lebar sepal (sepal_width) relatif lemah, 
#yang menunjukkan bahwa ukuran sepal tidak memiliki hubungan yang signifikan antara panjang dan lebarnya.
```
